```python
import os
from dotenv import load_dotenv

dotenv_path = '.env'
load_dotenv(dotenv_path)

SECRET_KEY = os.environ.get("APIKEY")
```

```python
import requests  

query_params = {
    "sources": "cnn",
    "from": "2021-03-11",
    "to": "2021-03-12",
    "language": "en",
    "sortBy": "top",
    "apiKey": SECRET_KEY
}

api_url="https://newsapi.org/v2/everything"
```

```python
res = requests.get(api_url, params=query_params)
res_json = res.json()
articles = res_json["articles"]
articles
```

```python
news_json = []

for i in range(0, len(articles), 1):
    curr=articles[i]
    date=curr["publishedAt"]
    title=curr["title"]
    full_article=curr["content"]
    author=curr["author"]
    news_json.append({
        "date": date,
        "title": title,
        "full_article": full_article,
        "author": author
    })
        
        
news_json
```

```python
import json

with open('news_json.json', 'w') as json_file:
    json.dump(news_json, json_file)
```
