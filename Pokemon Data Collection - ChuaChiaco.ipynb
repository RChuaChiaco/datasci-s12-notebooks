```python
import requests
URL="https://bulbapedia.bulbagarden.net/wiki/List_of_Pok%C3%A9mon_by_National_Pok%C3%A9dex_number"
page=requests.get(URL)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
poke_content=soup.find(id='mw-content-text')
poke_tables=poke_content.find_all('table')
```

```python
poke_list=poke_tables[2:9]
poke_list
```

```python
poke_json = []

for j in range(0, len(poke_list), 1):
    curr_list = poke_list[j]
    for i in range(3, len(curr_list.contents), 2):
        poke_info=curr_list.contents[i]
        kdex=poke_info.contents[1].text.strip()
        ndex=poke_info.contents[3].text.strip()
        name=poke_info.contents[7].text.strip()
        astring=str(poke_info.contents[7])
        url= "https://bulbapedia.bulbagarden.net" + astring.split('"')[1]
        type1=poke_info.contents[9].text.strip()
        gen=j+2
        if len(poke_info.contents) > 10:
            type2=poke_info.contents[11].text.strip()
            poke_json.append({
                "kdex": kdex,
                "ndex": ndex,
                "name": name,
                "type1": type1,
                "type2": type2,
                "url": url,
                "generation": gen
            })
        else:
            poke_json.append({
                "kdex": kdex,
                "ndex": ndex,
                "name": name,
                "type1": type1,
                "url": url,
                "generation": gen
            })
        
poke_json
```

```python
import json

with open('pkmn_gen_2_to_8.json', 'w') as json_file:
    json.dump(poke_json, json_file)
```
